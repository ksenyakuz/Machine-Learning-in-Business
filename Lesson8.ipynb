**ВОПРОС 2:**

Если мы увеличиваем объем выборки в два раза (при условии, что показатель стандартного отклонения остается неизменным), то 95% доверительный интервал
Выберите один вариант из списка

1) **Стал более узким**

2) Возможны оба варианта 

3) Стал более широким

В формуле расчета сигмы в знаменателе корень из n => при увеличении n, стандартное отклонение будет уменьшаться, поэтому с уменьшением сигмы интервал становится уже

```python
import numpy as np
import math

# Проверка
import statsmodels.api as sm

k = 479

for n in [1000, 2000]:
    print(
        f'Доверительный интервал, при n={n} ',
        sm.stats.proportion_confint(count=k, nobs=n, alpha=(1 - 0.95), method='normal'))
```

**ВОПРОС 3:**

В центре 95% доверительного интервала, рассчитанного по выборочным значениям, находится:
Выберите один вариант из списка

1) Значение стандартной ошибки среднего 

2) **Выборочное среднее значение** 

3) Среднее значение генеральной совокупности

```python
# Проверка
k, n = 479, 1000

interval = sm.stats.proportion_confint(k, n, alpha=0.05, method='normal')
np.median(interval) == (k / n)
```
